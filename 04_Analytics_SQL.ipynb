```python
spark.sql("""
CREATE OR REPLACE TEMP VIEW gold_readmissions
AS
SELECT * FROM delta.`/Volumes/workspace/default/pujitha_patient-readmission/gold`
""")
```

```python
%sql
SELECT
  COUNT(*) AS total_patients,
  SUM(readmitted_binary) AS readmitted_patients,
  ROUND(100.0 * SUM(readmitted_binary) / COUNT(*), 2) AS readmission_rate_percent
FROM gold_readmissions;
```

```python
%sql
SELECT
  admission_type,
  COUNT(*) AS total_cases,
  ROUND(100.0 * SUM(readmitted_binary) / COUNT(*), 2) AS readmission_rate
FROM gold_readmissions
GROUP BY admission_type
ORDER BY readmission_rate DESC;
```

```python
%sql
SELECT
  days_in_hospital,
  COUNT(*) AS total_cases,
  ROUND(100.0 * SUM(readmitted_binary) / COUNT(*), 2) AS readmission_rate
FROM gold_readmissions
GROUP BY days_in_hospital
ORDER BY days_in_hospital;
```
